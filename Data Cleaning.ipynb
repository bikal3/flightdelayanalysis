```python
import pandas as pd
import glob
```

```python
dtype_dict = {
    11: str,
    13: str,
    57: str,
    78: str,
    85: str,
    86: str,
    93: str
}
#Combinining All the dataset of 2022 in one csv.
# Folder path
folder_path = "Dataset/2022/"

# File extension
file_extension = "*.csv"

# Retrieve file paths
file_paths = glob.glob(folder_path + file_extension)

# Initialize an empty list to store the datasets
datasets = []

# Read and append each CSV file to the datasets list
for file_path in file_paths:
    try:
        dataset = pd.read_csv(file_path,dtype=dtype_dict)
        datasets.append(dataset)
    except pd.errors.EmptyDataError:
        print(f"Empty CSV file: {file_path}")

# Check if any datasets are found
if len(datasets) == 0:
    print("No datasets found.")

# Concatenate all datasets
dataset_2022 = pd.concat(datasets,ignore_index=True)

# Print the combined dataset
dataset_2022.head(10)
```
